```python
import pandas as pd
import numpy as np
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import Band, ColumnDataSource
output_notebook()
```

```python
d = pd.read_csv("E_test_cartpole.csv")
d
```

```python
Rm_1 = np.array(d["0.1"])
episodes_1 = np.array(range(len(Rm_1)))
mean_rewards_1 = []
std_rewards_1 = []
for i in range(len(Rm_1)):
    mean_rewards_1.append(np.mean(Rm_1[:(i+1)]))
    std_rewards_1.append(np.std(Rm_1[:(i+1)]))
mean_rewards_1 = np.array(mean_rewards_1)
std_rewards_1 = np.array(std_rewards_1)

Rm_2 = np.array(d["0.2"])
episodes_2 = np.array(range(len(Rm_2)))
mean_rewards_2 = []
std_rewards_2 = []
for i in range(len(Rm_2)):
    mean_rewards_2.append(np.mean(Rm_2[:(i+1)]))
    std_rewards_2.append(np.std(Rm_2[:(i+1)]))
mean_rewards_2 = np.array(mean_rewards_2)
std_rewards_2 = np.array(std_rewards_2)

Rm_3 = np.array(d["0.3"])
episodes_3 = np.array(range(len(Rm_3)))
mean_rewards_3 = []
std_rewards_3 = []
for i in range(len(Rm_3)):
    mean_rewards_3.append(np.mean(Rm_3[:(i+1)]))
    std_rewards_3.append(np.std(Rm_3[:(i+1)]))
mean_rewards_3 = np.array(mean_rewards_3)
std_rewards_3 = np.array(std_rewards_3)

Rm_4 = np.array(d["0.4"])
episodes_4 = np.array(range(len(Rm_4)))
mean_rewards_4 = []
std_rewards_4 = []
for i in range(len(Rm_4)):
    mean_rewards_4.append(np.mean(Rm_4[:(i+1)]))
    std_rewards_4.append(np.std(Rm_4[:(i+1)]))
mean_rewards_4 = np.array(mean_rewards_4)
std_rewards_4 = np.array(std_rewards_4)

Rm_5 = np.array(d["0.5"])
episodes_5 = np.array(range(len(Rm_5)))
mean_rewards_5 = []
std_rewards_5 = []
for i in range(len(Rm_5)):
    mean_rewards_5.append(np.mean(Rm_5[:(i+1)]))
    std_rewards_5.append(np.std(Rm_5[:(i+1)]))
mean_rewards_5 = np.array(mean_rewards_5)
std_rewards_5 = np.array(std_rewards_5)

Rm_6 = np.array(d["0.5"])
episodes_6 = np.array(range(len(Rm_5)))
mean_rewards_6 = []
std_rewards_6 = []
for i in range(len(Rm_6)):
    mean_rewards_6.append(np.mean(Rm_6[:(i+1)]))
    std_rewards_6.append(np.std(Rm_6[:(i+1)]))
mean_rewards_6 = np.array(mean_rewards_6)
std_rewards_6 = np.array(std_rewards_6)

Rm_7 = np.array(d["0.5"])
episodes_7 = np.array(range(len(Rm_5)))
mean_rewards_7 = []
std_rewards_7 = []
for i in range(len(Rm_7)):
    mean_rewards_7.append(np.mean(Rm_7[:(i+1)]))
    std_rewards_7.append(np.std(Rm_7[:(i+1)]))
mean_rewards_7 = np.array(mean_rewards_7)
std_rewards_7 = np.array(std_rewards_7)

Rm_8 = np.array(d["0.5"])
episodes_8 = np.array(range(len(Rm_8)))
mean_rewards_8 = []
std_rewards_8 = []
for i in range(len(Rm_8)):
    mean_rewards_8.append(np.mean(Rm_8[:(i+1)]))
    std_rewards_8.append(np.std(Rm_8[:(i+1)]))
mean_rewards_8 = np.array(mean_rewards_8)
std_rewards_8 = np.array(std_rewards_8)

Rm_9 = np.array(d["0.5"])
episodes_9 = np.array(range(len(Rm_8)))
mean_rewards_9 = []
std_rewards_9 = []
for i in range(len(Rm_9)):
    mean_rewards_9.append(np.mean(Rm_9[:(i+1)]))
    std_rewards_9.append(np.std(Rm_9[:(i+1)]))
mean_rewards_9 = np.array(mean_rewards_9)
std_rewards_9 = np.array(std_rewards_9)

source1 = ColumnDataSource(dict(
    x = episodes_1,
    mean = mean_rewards_1,
    std = std_rewards_1,
    upper = mean_rewards_1 + std_rewards_1,
    lower = mean_rewards_1 - std_rewards_1
))

source2 = ColumnDataSource(dict(
    x = episodes_2,
    mean = mean_rewards_2,
    std = std_rewards_2,
    upper = mean_rewards_2 + std_rewards_2,
    lower = mean_rewards_2 - std_rewards_2
))

source3 = ColumnDataSource(dict(
    x = episodes_3,
    mean = mean_rewards_3,
    std = std_rewards_3,
    upper = mean_rewards_3 + std_rewards_3,
    lower = mean_rewards_3 - std_rewards_3
))

source4 = ColumnDataSource(dict(
    x = episodes_4,
    mean = mean_rewards_4,
    std = std_rewards_4,
    upper = mean_rewards_4 + std_rewards_4,
    lower = mean_rewards_4 - std_rewards_4
))

source5 = ColumnDataSource(dict(
    x = episodes_5,
    mean = mean_rewards_5,
    std = std_rewards_5,
    upper = mean_rewards_5 + std_rewards_5,
    lower = mean_rewards_5 - std_rewards_5
))

p1 = figure(
    title = "Population Coding: Cartpole-v0",
    x_axis_label = "Episodes",
    y_axis_label = "Mean Rewards"
)

band1 = Band(base='x', lower='lower', upper='upper', source=source1, level='underlay',
            fill_alpha=0.2, line_width=1, line_color='black', fill_color='green')
p1.line(x=episodes_1, y=mean_rewards_1, line_width=2, color='green', legend_label='E = 0.1')

band2 = Band(base='x', lower='lower', upper='upper', source=source2, level='underlay',
            fill_alpha=0.2, line_width=1, line_color='black', fill_color='red')
p1.line(x=episodes_2, y=mean_rewards_2, line_width=2, color='red', legend_label='E = 0.2')

band3 = Band(base='x', lower='lower', upper='upper', source=source3, level='underlay',
            fill_alpha=0.2, line_width=1, line_color='black', fill_color='orange')
p1.line(x=episodes_3, y=mean_rewards_3, line_width=2, color='orange', legend_label='E = 0.3')

band4 = Band(base='x', lower='lower', upper='upper', source=source4, level='underlay',
            fill_alpha=0.2, line_width=1, line_color='black', fill_color='blue')
p1.line(x=episodes_4, y=mean_rewards_4, line_width=2, color='blue', legend_label='E = 0.4')

band5 = Band(base='x', lower='lower', upper='upper', source=source5, level='underlay',
            fill_alpha=0.2, line_width=1, line_color='black', fill_color='purple')
p1.line(x=episodes_5, y=mean_rewards_5, line_width=2, color='purple', legend_label='E = 0.5')

# p1.add_layout(band1)
# p1.add_layout(band2)
# p1.add_layout(band3)
# p1.add_layout(band4)
# p1.add_layout(band5)

p1.legend.location = 'bottom_right'

show(p1)
```
